# Min Coin
มีเหรียญ 1,2,5 บาท
หากมีจำนวนเงิน 49 บาท
จะถือเงินน้อยที่สุดกี่เหรียญ

```
#MinCoin
import pandas as pd
import numpy as np
```

```
def minCoin(x):
  if x < 0:
    return 0
  if x == 0:
    return 0
  if myDict[x] != 0:
    return myDict[x]
  else:
    c1 = x - 1
    cum1 = minCoin(c1) + 1 if c1 >= 0 else 100
    c2 = x - 2
    cum2 = minCoin(c2) + 1 if c2 >= 0 else 100
    c5 = x - 5
    cum5 = minCoin(c5) + 1 if c5 >= 0 else 100
  #print(x,cum1,cum2,cum5)
  myDict[x] = minSelect(cum1,cum2,cum5)
  return myDict[x]
def minSelect(x,y,z):
  return min(x,y,z)
```

# Chocolate Factory
ร้านขายขนม ทำขนมมา 2 สูตร
สูตร 1  ใช้ Chocolate 1 unit , นม 3 unit ได้กำไร 6 บาท
สูตร 2  ใช้ Chocolate 1 unit , นม 2 unit ได้กำไร 5 บาท
หากมี นม 12 unit และ Chocolate 5 unit 
ทำยังไงให้ได้กำไรมากที่สุด

```
#Chocolate
#Formula 1 1 chocolate,3 milk = 6 bath
#Formula 2 1 chocolate,2 milk = 5 bath
#We have 5 chocolate and 12 milk
def maxBath(c,m):
  if c <= 0 or m <= 0:
    return 0
  if c >= 1 and m >= 3:
    c1 = c-1
    m1 = m-3
    cum1 = maxBath(c1,m1) + 6
  else:
    cum1 = 0

  if c >= 1 and m >= 2:
    c2 = c-1
    m2 = m-2
    cum2 = maxBath(c2,m2) + 5
  else:
    cum2 = 0
  
  myDict[c][m] = maxSelect(cum1,cum2)
  return myDict[c][m]

def maxSelect(x,y):
  return max(x,y)
```

```
c = 5
m = 12
myDict = np.zeros([c+1,m+1])
print(maxBath(5,12))
#print(myDict)
```
